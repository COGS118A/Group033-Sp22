# Background

Natural language processing is a machine learning method that aims to help machines understand the human language. Clinical documents, drug reviews, physician reviews, etc., can be analyzed using sentiment analysis and by attempting to gain an understanding of the overall sentiment of a drug review and rating, we can begin to study the best course of treatment for different individual patients.

Sentiment analysis using natural language processing can be used to study the emotional/mental states and the subjective information provided by the writer. Sentiment analysis in a medical context has not been fully explored in a broader sense, but it is clear that it concerns patients' medical conditions, treatments, and overall patient health status<a name="sentiment"></a>[<sup>[1]</sup>](#sentimentnote). Sentiment analysis for medical documents and reviews require the context and understanding of the terminology, the names of the drugs and in what context they were used in, as well as the individual's specific meaning within that context. It is because of this complexity that machine learning techniques that try to determine the sentiment of a drug review need further study and improval in order to increase its generalizeability.

## More Data Processing
We first did some more cleaning of our data. Since we did not need much of the other columns for our model, we removed all of the columns other than the review and its rating. Then, we added the sentiment based off of the review by encoding 1-5 as being negative and 6-10 being positive. We then dropped the rating column as it was not needed anymore for our binary classification, and one-hot encoded positive as 1 and negative as 0.

```python
clean_df = full_df.drop(['drugName', 'condition', 'uniqueID', 'date', 'usefulCount'], axis=1)
```

```python
X = clean_df.review
y = clean_df.sentiment

# One-hot encode 'positive' -> 1, 'negative' -> 0
y = np.unique(y, return_inverse=True)[1]
y
```

## Count Vectorizer
Following our preprocessing, we then used a count vectorizer in order to translate the words so that the algorithm can decipher the input. This turns each review into a vector of word counts.

```python
cv = CountVectorizer()
pipe = Pipeline([('vect', cv)])

X_features = pipe.fit_transform(X)
```

```python
X_features.shape
y.shape
```

## Learning Curve
Now we needed to find our optimal split for our training and testing data. We used a learning curve in order to find this, and found that our optimal split was 80/20 train/test.

```python
# Plotting learning curve
model = LogisticRegression(solver='saga', random_state=0, 
                           max_iter=4000, verbose=3)

train_size, train_score, test_score = learning_curve(model, X_features, y, train_sizes=[106349, 127618, 148888, 170158])
```

```python
plt.plot(train_size, np.mean(train_score, axis=1))
plt.plot(train_size, np.mean(test_score, axis=1))
```

## Logistic Regression
We chose to use Logistic regression as our base model using a pipeline that utilizes a count vectorizer to create our train test split. The model ended up with an accuracy score of .856 which although isn't the best is still high enough to be used as a reliable model. The learning curve shows that we could use an 80/20 train/test split for the best accuracy percentage using a 5-fold cross-validation for validation.

```python
X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.2, random_state=42)
```

```python
# Baseline model
model = LogisticRegression(solver='saga', random_state=0, 
                           max_iter=4000, verbose=3)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

## Gridsearch Cross Validation
Continuing to use Logistic Regression as our base model, we conducted a Gridsearch Cross Validation(GSCV) that has 5 folds, using 6 different values of C starting from 0.0001 and multiplies by 10 until C is 1 and 2 different solvers, "sag" and "saga" having a total of 12 candidates, with 60 different fits. The model with the highest score had the same hyperparameters as base model being at, .85, with C = 1 and solver = "saga".

```python
# Gridserch CV for best hyperparams:

gs_model = LogisticRegression(random_state=0, max_iter=4000)
gscv = GridSearchCV(gs_model, param_grid={'C': [0.0001, 0.001, 0.01, 0.01, 0.1, 1], 'solver': ['sag', 'saga']}, verbose=3)
gscv.fit(X_train, y_train)
```

```python
gscv.cv_results_['mean_test_score']
```

## ROC Curve 
We then found the True Positive Rate (TPR) and the False Positive Rate (FPR) and created a Receiver Operator Characteristic (ROC) Curve. The ROC Curve helps prove our model to be accurate as it has has a high Area Under the Curve(AOC) at 0.91 and approaches TPR = 1 very quickly.

```python
# C=1, solver='saga' is the best in terms of avg. acc %

predictions = gscv.predict_proba(X_test)[:,1]

roc_auc = roc_auc_score(y_test, predictions)
fpr, tpr, _ = roc_curve(y_test, predictions)
```

```python
plt.figure(dpi=100)
plt.plot(fpr, tpr)
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

print('Area under the Receiver Operating Characteristic curve:', 
      roc_auc)
```
